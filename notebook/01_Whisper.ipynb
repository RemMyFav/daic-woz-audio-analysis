```python
from pathlib import Path
from tqdm.auto import tqdm
import pandas as pd
from datetime import datetime
PROJECT_ROOT = Path.cwd().parent
import whisper
import sys
import torch
sys.path.append(str(PROJECT_ROOT))
from src.asr.whisper import run_whisper_pipeline
```

```python
PROJECT_ROOT = Path.cwd().parent

MODEL_NAME = "base"

WHISPER_OUTPUT_DIR = (
    PROJECT_ROOT
    / "outputs"
    / "asr"
    / f"whisper_{MODEL_NAME}"
)

LOG_DIR = PROJECT_ROOT / "logs"

SUCCESS_LOG = (
    LOG_DIR
    / f"whisper_{MODEL_NAME}_success.csv"
)

FAILED_LOG = (
    LOG_DIR
    / f"whisper_{MODEL_NAME}_failures.csv"
)

RAW_DIR = PROJECT_ROOT / "data" / "raw"
AUDIO_DIR = PROJECT_ROOT / "data" / "audio"
TRANSCRIPT_DIR = PROJECT_ROOT / "data" / "transcript"

for directory in [
    RAW_DIR,
    AUDIO_DIR,
    TRANSCRIPT_DIR,
    WHISPER_OUTPUT_DIR,
    LOG_DIR,
]:
    directory.mkdir(
        parents=True,
        exist_ok=True,
    )

audio_files = sorted(
    AUDIO_DIR.glob("*_AUDIO.wav")
)

print(
    f"Found {len(audio_files)} audio files"
)
```

```python
device = (
    "cuda"
    if torch.cuda.is_available()
    else "cpu"
)

model = whisper.load_model(
    MODEL_NAME,
    device=device,
)

print(
    f"Model : {MODEL_NAME}"
)

print(
    f"Device: {model.device}"
)
```

```python
run_whisper_pipeline(
    model=model,
    audio_paths=audio_files,
    output_dir=WHISPER_OUTPUT_DIR,
    success_log=SUCCESS_LOG,
    failure_log=FAILED_LOG,
    language="en",
    overwrite=False,
)
```
