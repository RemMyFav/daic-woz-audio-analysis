```python
from pathlib import Path
from tqdm.auto import tqdm
import pandas as pd
from datetime import datetime
PROJECT_ROOT = Path.cwd().parent
import whisper
import sys
sys.path.append(str(PROJECT_ROOT))
from src.asr.whisper import transcribe_audio
```

```python
WHISPER_OUTPUT_DIR = PROJECT_ROOT / "outputs" / "whisper"
LOG_DIR = PROJECT_ROOT / "logs"
SUCCESS_LOG = LOG_DIR / "whisper_success.csv"
FAILED_LOG = LOG_DIR / "whisper_failures.csv"
RAW_DIR = PROJECT_ROOT / "data" / "raw"
AUDIO_DIR = PROJECT_ROOT / "data" / "audio"
TRANSCRIPT_DIR = PROJECT_ROOT / "data" / "transcript"

for directory in [
    RAW_DIR,
    AUDIO_DIR,
    TRANSCRIPT_DIR,
    WHISPER_OUTPUT_DIR,
    LOG_DIR,
]:
    directory.mkdir(
        parents=True,
        exist_ok=True,
    )


audio_files = sorted(AUDIO_DIR.glob("*_AUDIO.wav"))
```

```python
model = whisper.load_model("base")
```

```python
audio_files = sorted(
    AUDIO_DIR.glob("*_AUDIO.wav")
)

success_count = 0
failed_count = 0

for audio_path in tqdm(
    audio_files,
    desc="Running Whisper",
    unit="audio",
):

    try:

        result = transcribe_audio(
            model=model,
            audio_path=audio_path,
            output_dir=WHISPER_OUTPUT_DIR,
        )

        success_row = pd.DataFrame([
            {
                "data_id": result["data_id"],
                "csv_path": str(
                    result["csv_path"]
                ),
                "timestamp": datetime.now().isoformat(),
            }
        ])

        success_row.to_csv(
            SUCCESS_LOG,
            mode="a",
            header=not SUCCESS_LOG .exists(),
            index=False,
        )

        success_count += 1

        tqdm.write(
            f"[SUCCESS] "
            f"{result['data_id']} "
        )

    except Exception as e:

        failed_row = pd.DataFrame([
            {
                "audio_file": str(audio_path),
                "error_type": type(e).__name__,
                "error": str(e),
                "timestamp": datetime.now().isoformat(),
            }
        ])

        failed_row.to_csv(
            FAILED_LOG,
            mode="a",
            header=not FAILED_LOG.exists(),
            index=False,
        )

        failed_count += 1

        tqdm.write(
            f"[FAILED] "
            f"{audio_path.name} | "
            f"{type(e).__name__}: {e}"
        )

print("\n====================")
print(f"Success : {success_count}")
print(f"Failed  : {failed_count}")
print("====================")
```
